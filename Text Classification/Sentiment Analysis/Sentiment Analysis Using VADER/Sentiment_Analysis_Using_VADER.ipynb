```python
ds = pd.read_excel('/content/Retail.xlsx')
ds.head()
```

```python
Vader_Df = pd.DataFrame()
Vader_Df['reviews'] = ds['reviews.text']
Vader_Df['Polarity_Scores'] = Vader_Df['reviews'].apply(lambda reviews: sentiment.polarity_scores(reviews))
Vader_Df['Compound_Score'] = Vader_Df['Polarity_Scores'].apply(lambda x: x['compound'])
```

```python
Vader_Df.head()
```

### Let's Bucket the text into different sentiment brackets.
- Very negative = `1 to -0.5` 
- Negative = `-0.5 to -0.2`
- Neutral `-0.2 to 0.2`  
- Positive `0.2 to 0.5`
- Very Positive `0.5 to 1`

```python
func = lambda x:'Very Negative' if (x >= -1) & (x < -0.5) \
else ('Negative' if (x >= -0.5) & (x < -0.2) \
else ('Neutral' if (x>=-0.2)&(x<0.2) \
else ('Positive' if (x>=0.2) & (x<0.5) \
else 'Very Positive')   ))
```

```python
Vader_Df['Sentiment'] = Vader_Df['Compound_Score'].apply(func)
```
